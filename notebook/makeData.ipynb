```python
idx = np.minimum(np.random.binomial(len(devices),0.4,1000000), len(devices)-1)
device_list = [devices[i] for i in idx]
# print(Counter(device_list))
Counter(device_list)
```

```python
asset_list = [asset[min(i, len(asset)-1)] for i in np.random.binomial(len(asset), 0.4, 1000000)]
Counter(asset_list)
```

```python
rbr_list = np.random.normal(1.0, 0.5,1000000)
rbr_list
```

```python
state_list = [state[i] for i in np.minimum(np.maximum(np.random.normal(len(state)/2, len(state)/5, 1000000).astype(int),0),51)]
Counter(state_list)
```

```python
has_ended = [(1 if np.random.rand()>0.4 else 0) if i>0.4 else 0 for i in np.random.rand(1000000)]
is_newSession = [1 if i>0.6 else 0 for i in np.random.rand(1000000)]
Counter(has_ended), Counter(is_newSession)
```

```python
is_vsf = [1 if i>0.985 else 0 for i in np.random.rand(1000000)]
is_ebvs = [1 if i>0.99 else 0 for i in np.random.rand(1000000)]
Counter(is_vsf), Counter(is_ebvs)
```

```python
session_length = (np.random.poisson(5, 1000000) * np.random.poisson(4, 1000000)) + np.random.rand(1000000)
ts = np.concatenate((np.random.uniform(0,21,900000).astype(int), np.random.poisson(4.5,50000), np.random.normal(13.5,1.5,50000).astype(int)))
Counter(ts)
```

```python
df = pd.DataFrame({'ts':np.minimum(ts,20),
                   'asset': asset_list,
                   'state': state_list,
                   'device': device_list,
                   'rbr': rbr_list,
                   'justJoined': is_newSession,
                   'hasEnded': has_ended,
                   'isVsf': is_vsf,
                   'isEbvs': is_ebvs,
                   'sessionLength':session_length})
df['hasEnded'] = np.maximum(np.maximum(df.hasEnded, df.isVsf), df.isEbvs)
df.loc[df.justJoined==1, 'sessionLength'] = df.loc[df.justJoined==1, 'sessionLength']-df.loc[df.justJoined==1, 'sessionLength'].astype(int)
df = df.sample(len(df))
# df['concurrentPlays'] = np.random.randint()
df[:5]
```

```python
import itertools
all_comb = pd.DataFrame(itertools.product(*[range(18),asset, state,devices]), columns = ['ts','asset','state','device'])
all_comb[:5]
```

```python
agg_df = df.groupby(['ts','asset','state','device']).agg({#'concurrentPlays':'sum',
                                                         'rbr':'mean', 
                                                         'justJoined':'sum',
                                                         'hasEnded':'sum',
                                                         'isVsf':'sum',
                                                         'isEbvs':'sum',
                                                         'sessionLength':'mean'}).reset_index()
agg_df = pd.merge(all_comb, agg_df, how='left')
agg_df['justJoined'] = agg_df.justJoined.fillna(0)
agg_df['hasEnded'] = agg_df.hasEnded.fillna(0)
agg_df['isVsf'] = agg_df.isVsf.fillna(0)
agg_df['isEbvs'] = agg_df.isEbvs.fillna(0)

agg_df['sessionLength'] = agg_df.groupby(['asset','state','device'])['sessionLength'].transform(lambda x: x.interpolate().bfill().fillna(6.2))
agg_df['rbr'] = np.abs(agg_df.groupby(['asset','state','device'])['rbr'].transform(lambda x: x.interpolate().bfill().fillna(0.6)))

agg_df.loc[agg_df.ts==0, 'concurrentPlays'] = np.maximum((np.random.randint(1,6,(agg_df.ts==0).sum()).astype(int) * np.random.randint(0,7,(agg_df.ts==0).sum()).astype(int)) - np.random.randint(1,17,(agg_df.ts==0).sum()).astype(int),0)
agg_df.loc[agg_df.ts==0, 'hasEnded'] = [np.random.randint(i//10, i+1) for i in (agg_df.loc[agg_df.ts==0, 'concurrentPlays'].values)]

agg_df = agg_df.sort_values('ts').reset_index(drop=True)
for i in range(1,18):
    ith_lag = agg_df.groupby(['asset','state','device'])['concurrentPlays'].shift(1)
    ith_lag_end = agg_df.groupby(['asset','state','device'])['hasEnded'].shift(1)
    val = ith_lag.loc[agg_df.ts==i] -\
         ith_lag_end.loc[agg_df.ts==i] +\
         agg_df.loc[agg_df.ts==i,'justJoined']
    agg_df.loc[(agg_df.ts==i)&(val<0),'justJoined'] = agg_df.loc[(agg_df.ts==i)&(val<0),'justJoined'] -\
                                                        val[(agg_df.ts==i)&(val<0)]
    agg_df.loc[agg_df.ts==i,'concurrentPlays'] = ith_lag.loc[agg_df.ts==i] -\
                                                 ith_lag_end.loc[agg_df.ts==i] +\
                                                 agg_df.loc[agg_df.ts==i,'justJoined']
    agg_df.loc[agg_df.ts==i, 'hasEnded'] = [np.random.randint(i//4, i+1) for i in (agg_df.loc[agg_df.ts==i, 'concurrentPlays'].values)]
    
agg_df[:5]
```
